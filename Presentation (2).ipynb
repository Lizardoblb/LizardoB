# Description of the Data Sheet

* This datasheet contains information about the percentage of the rural population in each country from the 1960 to the 2020, including their name and country code.

* Country : Specifies the name of each country, this being a categorical variable

* Rows 1960 - 2020 : Represents each years migration percentage, these are numeric variables

* Link to spreadsheet: https://data.worldbank.org/indicator/SP.RUR.TOTL.ZS?view=chart

* Data analyzed by the World Bank staff based on the United Nations Population Division's World Urbanization Prospects: 2018 Revision.

* Questions and observations

Governments and companies would be very interested in this data because they could estimate the growth and job output required for their countries, governments regarding social issues and needs, and companies could analyze for possible viability due to the city growth.

# Hows the similarity in migration to urban areas between Mexico and the United States?

```python
#print(df.loc['Mexico'])
graph = plt.figure()
graph.set_figwidth(35)
graph.set_figheight(5)
plt.plot(df.loc['Mexico','1960':'2020'])
plt.xlabel('Years', fontsize = 23)
plt.ylabel('% of rural Pop', fontsize = 17)
plt.title('Mexico', fontsize = 30)
plt.show()
graph = plt.figure()
graph.set_figwidth(35)
graph.set_figheight(5)
plt.plot(df.loc['United States','1960':'2020'])
plt.xlabel('Years', fontsize = 23)
plt.ylabel('% of rural Pop', fontsize = 17)
plt.title('United States', fontsize = 30)
plt.show()
```

* For this first question we will analyze how Mexico and its northern neighbor look like in terms of rural pop percentage.
  We can observe that in 1960, Mexico has a 50% rural pop and the US has a 30%, till this day the percentage is pretty most the same, 
  leaving Mexico in 2020 with a 19.263% and the US with a 17.3%.

# How has the migration from rural to urban pop changed in Mexico in the last 15 years?

```python
graph = plt.figure()
graph.set_figwidth(35)
graph.set_figheight(5)
plt.plot(df.loc['Mexico','2005':'2020'])
plt.xlabel('Years', fontsize = 23)
plt.xticks( fontsize = 17)
plt.ylabel('% of rural Pop', fontsize = 17)
plt.title('Mexico', fontsize = 30)
plt.show()
```

* After analyzing last graphs, i noticed that Mexico had a decently decadent line in terms of its rural poulation, so with this information 
  from 2015 till today, we can assume that the rural pop % goes down aproximately 0.333% per year, making this a useful forecast for possible 
  companies in rural or urban por areas
