```python
df = pd.read_excel('datasets/mic2.xlsx',engine = 'openpyxl')
df
```

```python
mpl.style.use('ggplot')
```

```python
counts = df['Age'].value_counts()
counts
```

```python
counts = counts.reindex(['<=20','21 - 30','31 - 40','41 - 50'])
counts
```

```python
type(counts)
counts = counts.to_frame()
```

```python
counts['%'] = (counts.iloc[:, :]/counts.loc[:,'Age'].sum(axis = 0))*100
```

```python
counts
```

```python
counts = counts.reset_index()
counts
```

```python
counts = counts.rename({'index':'Age_Group'}, axis = 1)
```

```python
counts
```

```python
counts['Age'].plot(kind = 'pie')
```

```python
x = list(counts['Age_Group'])
y = list(counts['Age'])
plt.subplot(1,2,1)
plt.rcParams['figure.figsize'] = (15,7)
plt.bar(x,y,width = 0.7, color = 'green')

plt.subplot(1,2,2)
explode_list = [0.08, 0.08, 0.08, 0.08]
color_list = ['gold', 'red', 'coral', 'skyblue', 'lightgreen', 'red', 'purple']
counts['Age'].plot(kind = 'pie',
                  figsize = (15,6),
                  autopct = '%1.1f%%',
                    startangle = 90,
                   shadow = True,
                   pctdistance = 1.1,
                   labels = None,
                   explode = explode_list,
                   colors = color_list
                  )
```
