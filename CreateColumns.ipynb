```python
fd = pd.read_csv('datasets/cwurData.csv')
fd.head(1)
```

```python
def create_category(x):
    if (x >= 1) & (x<= 100):
        return 'First Tier'
    elif (x > 100) & (x <=200):
        return 'Second Tier'
    elif (x > 200):
        return 'Other Top University'
```

```python
fd['Rank_level'] = fd['world_rank'].apply(lambda y: create_category(y))
fd
```
