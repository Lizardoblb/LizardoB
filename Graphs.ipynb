```python
import seaborn as sns
```

```python
%ls datasets/'Cartwheeldata.csv'
```

```python
df = pd.read_csv('datasets/Cartwheeldata.csv')
df.head()
```

```python
sns.lmplot(x = 'Wingspan', y = 'CWDistance', data = df, hue = 'Gender')
plt.title('Scatter plot of CWDistance vs Wingspan', fontsize = 10)
plt.show()
```

```python
sns.swarmplot(x = 'Gender', y = 'CWDistance', data = df)
```

```python
sns.boxplot (data = df.loc[df['Gender']=='M',['Age', 'Height', 'Wingspan', 'CWDistance', 'Score']])
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(1,2,1) #(rows, columns, positions
sns.boxplot(data = df.loc[df['Gender']=='M', ['Score', 'Age']])
ax2 = fig.add_subplot(1,2,2) #(rows, columns, positions
sns.boxplot(data = df.loc[df['Gender']=='F', ['Score', 'Age']])
ax1.title.set_text('Male box plot')
ax2.title.set_text('Female box plot')
plt.show()
```

```python
sns.distplot(df.CWDistance)
plt.show()
```

```python
sns.countplot(x = 'Gender', data = df)
```
