## Description of datasest

### This database indicates in % the decline of rural population in the different states of the USA. Country Name represents the state, Country code the state code, Indicator Name indicates whether it is a rural population and Indicator Code is the code of the type of population.

```python
df = pd.read_excel('Rural Populations2.xlsx',
                   sheet_name = 'Hoja1',
                   engine = 'openpyxl',
                   skiprows = range(0),
                   skipfooter = 2
                  )
df.head(2)
```

## When viewing the database in excel and filtering the data, I realized that 'Indicator Name' and 'Indicator Code' have the same data in all their cells, so I decided to eliminate these 2 columns.

```python
df.columns
```

```python
df.drop(['Indicator Code','Indicator Name',],
        axis = 1,
        inplace = True)
df.columns
```

## I decided to abbreviate the name of the remaining 2 columns


```python
df.rename(columns = {'Country Name' : 'CN',  'Country Code' : 'CC', },
          inplace = True
         )
```

```python
df.head()
```

```python
df.columns = list(map(str,df.columns))
df.columns
```

## In order to identify the states and compare them with the years, it was necessary to separate 'CN' from the original table


```python
df.set_index('CN', inplace = True)
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
df.head(2)
```

## What was the % of rural population decreased in Germany from 1976 to 2016?


```python
df.loc['Germany', '1976' :].plot()
```

## Compare the % of rural population decreased in Germany and in Costa Rica since the same year


```python
ci = df.loc[['Germany','Costa Rica'], '1980':].transpose()
ci.plot()
```

## As we can see in Costa Rica there was a more severe drop in rural population over the years compared to Germany
