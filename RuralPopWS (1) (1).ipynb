```python
for number in range(154,155):
    print(df.iloc[number,0])
    y2 = df.iloc[number,4:65]
    graph = plt.figure()
    graph.set_figwidth(35)
    graph.set_figheight(5)
    plt.plot(x,y2)
    plt.xlabel('Years')
    plt.ylabel('% of rural Pop')
    plt.show()
```
